```python
#importing libraries
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle= 90, autopct='%1.1f%%')
plt.title('Work from home')
plt.ylabel('')
plt.show()
```

```python
#plotting these data that have boolean values into pie charts
df[['job_work_from_home', 'job_no_degree_mention', 'job_health_insurance']]
```

```python
fig, ax = plt.subplots(1,3)

#first we are plotting job work from home using matplotlib
#we will make a loop so we don't need to write our code every time but first we provide a dictionary

dict_column = {
    'job_work_from_home': 'Work from home', #value is what it will be written in our chart
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance' : 'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()): #because of dictionary value we are providing column, title
    ax[i].pie(df[column].value_counts(),startangle= 90, autopct='%1.1f%%', labels=['False', 'True'])
    ax[i].set_title(title)
```

```python
#when we start this code we can see it bring index and touple of key and value
for i, (column, title) in enumerate(dict_column.items()):
    print(i, (column,title))
```

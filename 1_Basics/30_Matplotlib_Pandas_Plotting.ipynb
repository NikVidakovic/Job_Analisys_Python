```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_counts = df.job_title_short.value_counts()
```

```python
job_counts = job_counts.sort_values(ascending=False) #sorting values so chart can look better
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45,  ha='right') #making x axis names lined up
plt.show()
```

```python
job_counts.plot(kind='bar')
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Count of Job Postings')
plt.xlabel('')
plt.xticks(rotation=45,  ha='right') #making x axis names lined up
plt.show()
```

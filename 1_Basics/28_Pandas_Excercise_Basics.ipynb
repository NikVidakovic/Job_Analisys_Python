```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y)
plt.show()
```

```python
#plotting job postings date
df.job_posted_date.head()
```

```python
plt.plot(df.job_posted_date, df.job_posted_date)
```

```python
date_counts = df.job_posted_date.value_counts() #left side is the index
date_counts
```

```python
date_counts = date_counts.sort_index()
```

```python
plt.plot(date_counts.index, date_counts)
```

```python
#adding a new column in which we are accessing month number 
df['job_posted_month'] = df['job_posted_date'].dt.month
df
```

```python
monthly_counts = df.job_posted_month.value_counts()
monthly_counts = monthly_counts.sort_index()
monthly_counts
```

```python
plt.plot(monthly_counts.index, monthly_counts.values)
```

```python
job_counts = df.job_title_short.value_counts()
```

```python
job_counts #indexes are names 
```

```python
job_counts = job_counts.sort_values(ascending=False) #sorting values so chart can look better
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45,  ha='right') #making x axis names lined up
plt.show()
```

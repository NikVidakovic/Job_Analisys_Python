# How are in-demand skills trending for Data Analysts?

### Methodology

1. Aggregate skill counts monthly
2. Re-analyze based on percentage of total jobs
3. Plot the monthly skill demand

```python
#importing libraries
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#filtering data analyst and US
df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title'] == 'Data Analyst')].copy()
```

```python
#adding month no column
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
#pivoting data so we can get count and also filling NaN values with 0
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

#addig total column at the bottom 
df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

#we need to sort by ro so we are filtering dataframe - specifying indexbecuase it is a series
df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

#droping total row so we can plot
df_DA_US_pivot = df_DA_US_pivot.drop('Total')
df_DA_US_pivot
```

```python
#getting a percantage of demanded skills for data analysts in US but first we get a number of job postings in every month 
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
#now we need to divide every row by totals so we can get a percentage
df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)
```

```python
#Changing month number to month name
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')
df_DA_US_percent
```

```python
import seaborn as sns
#selecting all rows but only 5 columns
df_plot = df_DA_US_percent.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()
plt.title('Trending Top Skills for Data Analysts in the US')
plt.xlabel('2023')
plt.ylabel('Likelyhood in Job Postings')
plt.legend().remove()

#formatting y axis so we can have % labels
from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

#adding text labels so we are making a loop
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1,i], df_plot.columns[i]) #11 is for provinding x axis, y is the last index in df_plot (x), (y) is index

plt.show()
```
